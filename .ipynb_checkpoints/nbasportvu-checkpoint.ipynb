```python
df = df.reset_index(drop=True)
```

```python
df_copy = df.copy()
```

```python
df['SHOT_CLOCK'] = df['SHOT_CLOCK'].astype(float)
df = df[np.isfinite(df['SHOT_CLOCK'])]
```

```python
df_shotclock = df[['SHOT_CLOCK', 'SHOT_RESULT']].copy()
```

```python
df_shotclock.SHOT_CLOCK = df_shotclock.SHOT_CLOCK.apply(round)
```

```python
def convert(s):
    if s == 'made':
        return 1
    return 0
```

```python
df_shotclock.SHOT_RESULT
```

```python
df_shotclock.SHOT_RESULT = df_shotclock.SHOT_RESULT.apply(convert)
```

```python
df_clockgrouped = df_shotclock.groupby('SHOT_CLOCK').aggregate([sum, len])
```

```python
shotclockresults = df_clockgrouped.SHOT_RESULT['sum']/df_clockgrouped.SHOT_RESULT['len']
shotclockresults 
```

```python
fig = plt.plot(df_clockgrouped.SHOT_RESULT['len'])
plt.xlabel('Shot clock remaining (sec)')
plt.ylabel('FGA')
plt.title('FGA vs. Shot Clock')

fig = plt.gcf()

plot_url = py.plot_mpl(fig, filename='mpl-basic-line')
```
