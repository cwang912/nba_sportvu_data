```python
df_shotclock = df[['SHOT_CLOCK', 'SHOT_RESULT']].copy()
df_shotclock.SHOT_CLOCK = df_shotclock.SHOT_CLOCK.apply(round)

def convert(s):
    if s == 'made':
        return 1
    return 0

df_shotclock.SHOT_RESULT = df_shotclock.SHOT_RESULT.apply(convert)

df_clockgrouped = df_shotclock.groupby('SHOT_CLOCK').aggregate([sum, len])
shotclockresults = df_clockgrouped.SHOT_RESULT['sum']/df_clockgrouped.SHOT_RESULT['len']

plt.style.use('fivethirtyeight')
plt.plot(shotclockresults)
plt.xlabel('Shot clock remaining (sec)')
plt.ylabel('FG %')
plt.title('FG % by time',x=0.05,y=1.1)
ax = plt.gca()
ax.invert_xaxis()
```

```python
# FILTER BY AN INDIVIDUAL
def fgp_vs_shotclock_ind(firstname, lastname, df):
    fig = plt.figure(figsize=(8,6))
#     majorFormatter = matplotlib.ticker.FormatStrFormatter('%d')

    def to_percent(y, position):
        # Ignore the passed in position. This has the effect of scaling the default
        # tick locations.
        s = str(round(100 * y))[:-2]
        # The percent symbol needs escaping in latex
        if matplotlib.rcParams['text.usetex'] == True:
            return s + r'$\%$'
        else:
            return s + '%'

    # Create the formatter using the function to_percent. This multiplies all the
    # default labels by 100, making them all percentages
    formatter = matplotlib.ticker.FuncFormatter(to_percent)

    # Set the formatter
    plt.gca().yaxis.set_major_formatter(formatter)

    plt.style.use('fivethirtyeight')
    
    def convert(s):
        if s == 'made':
            return 1
        return 0

    df_shotclock = df[['SHOT_CLOCK', 'SHOT_RESULT']].copy()
    df_shotclock.SHOT_CLOCK = df_shotclock.SHOT_CLOCK.apply(round)
    df_shotclock.SHOT_RESULT = df_shotclock.SHOT_RESULT.apply(convert)
    df_clockgrouped = df_shotclock.groupby('SHOT_CLOCK').aggregate([sum, len])
    shotclockresults = df_clockgrouped.SHOT_RESULT['sum']/df_clockgrouped.SHOT_RESULT['len']
    
    plt.plot(shotclockresults)
    
    name = lastname + ', ' + firstname
    df_ind = df[df.SHOOTER_NAME==name]
    df_shotclock_ind = df_ind[['SHOT_CLOCK', 'SHOT_RESULT']].copy()
    df_shotclock_ind.SHOT_CLOCK = df_shotclock_ind.SHOT_CLOCK.apply(round)
    df_shotclock_ind.SHOT_RESULT = df_shotclock_ind.SHOT_RESULT.apply(convert)
    df_clockgrouped_ind = df_shotclock_ind.groupby('SHOT_CLOCK').aggregate([sum, len])
    shotclockresults_ind = df_clockgrouped_ind.SHOT_RESULT['sum']/df_clockgrouped_ind.SHOT_RESULT['len']

    

#     ax.yaxis.set_major_formatter(majorFormatter)
    
    plt.plot(shotclockresults_ind, 'red')
    plt.xlabel('Shot clock remaining (sec)')
    plt.ylabel('Field goal percentage', x=0,y=0.5)
    plt.title('Field goal percentage by time, 2013-2014',x=0.3,y=1.1)
    plt.ylim(0,1)
    plt.xlim(24,0)
    plt.legend(['League average',firstname +' '+ lastname],'upper right')
    fig.tight_layout()
#     fig = plt.plots() 
    fig.savefig(firstname+lastname+'_FGPvsTime.eps', format='eps',bbox_indches='tight', pad_inches=1)
```
